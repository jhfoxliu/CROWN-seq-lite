# CROWN-seq-lite example mapping pipeline

## 1. Cutadapt

Note: 1 million test reads

```python

!cutadapt -m 32 -U 1 -j 20 -q 20 -e 0.25 -a NAGATCGGAAGAGCACACGTC -A TAAN{{11}}AGATCGGAAGAGCGTCGTG -o Ctrl_R1.cutadapt.fastq -p Ctrl_R2.cutadapt.fastq Ctrl_R1.fastq.gz Ctrl_R2.fastq.gz
```

```python
!cutadapt -m 32 -U 1 -j 20 -q 20 -e 0.25 -a NAGATCGGAAGAGCACACGTC -A TAAN{{11}}AGATCGGAAGAGCGTCGTG -o FtoKO_R1.cutadapt.fastq -p FtoKO_R2.cutadapt.fastq FtoKO_R1.fastq.gz FtoKO_R2.fastq.gz
```

## 2. UMI-tools barcodes

```python
!umi_tools extract -p NNNNNNNNNNNNNN -I Ctrl_R1.cutadapt.fastq -S Ctrl_R1.umi.fastq --read2-in Ctrl_R2.cutadapt.fastq --read2-out Ctrl_R2.umi.fastq
```

```python
!umi_tools extract -p NNNNNNNNNNNNNN -I FtoKO_R1.cutadapt.fastq -S FtoKO_R1.umi.fastq --read2-in FtoKO_R2.cutadapt.fastq --read2-out FtoKO_R2.umi.fastq
```

## 3. A-to-G conversion

```python
!python fastq_a2g.py -i Ctrl_R1.umi.fastq > Ctrl_R1.A2G.fastq
!python fastq_t2c.py -i Ctrl_R2.umi.fastq > Ctrl_R2.T2C.fastq
```

```python
!python fastq_a2g.py -i FtoKO_R1.umi.fastq > FtoKO_R1.A2G.fastq
!python fastq_t2c.py -i FtoKO_R2.umi.fastq > FtoKO_R2.T2C.fastq
```

## 4. Mapping to A-to-G converted transcriptome

```python
!bowtie2 -x ./Metadata/bowtie2_index/snRNA_snoRNA.A2G -1 Ctrl_R1.A2G.fastq -2 Ctrl_R2.T2C.fastq -p 20 --no-unal --norc --no-mixed --no-discordant -k 20 --end-to-end | samtools view -bS -F 4 > Ctrl.bam
```

```python
!bowtie2 -x ./Metadata/bowtie2_index/snRNA_snoRNA.A2G -1 FtoKO_R1.A2G.fastq -2 FtoKO_R2.T2C.fastq -p 20 --no-unal --norc --no-mixed --no-discordant -k 20 --end-to-end  | samtools view -bS -F 4 > FtoKO.bam
```

## 5. Deduplication

```python
!python Bam_recovery_A2G_PE.py -i Ctrl.bam -o Ctrl.recovered.bam -f Ctrl_R1.umi.fastq -r Ctrl_R2.umi.fastq
```

```python
!python Bam_recovery_A2G_PE.py -i FtoKO.bam -o FtoKO.recovered.bam -f FtoKO_R1.umi.fastq -r FtoKO_R2.umi.fastq
```

```python
!samtools sort -@ 4 -m 4G -o Ctrl.recovered.sorted.bam Ctrl.recovered.bam
!samtools index Ctrl.recovered.sorted.bam
```

```python
!samtools sort -@ 4 -m 4G -o FtoKO.recovered.sorted.bam FtoKO.recovered.bam
!samtools index FtoKO.recovered.sorted.bam
```

## 6. Deduplication

```python
!umi_tools dedup --paired --chimeric-pairs=discard --unpaired-reads=discard --stdin=Ctrl.recovered.sorted.bam --log=umi.logs --method=unique > Ctrl.dedeup.bam
!samtools index  Ctrl.dedeup.bam
```

```python
!umi_tools dedup --paired --chimeric-pairs=discard --unpaired-reads=discard --stdin=FtoKO.recovered.sorted.bam --log=umi.logs --method=unique > FtoKO.dedeup.bam
!samtools index  FtoKO.dedeup.bam
```

## 7. Get methylation data

```python
!python filter_snRNA_mapping_output.py -i Ctrl.dedeup.bam -o Ctrl.csv --db ./Metadata/snRNA_mouse_brain.database.txt --genelist  ./Metadata/gencode.vM35.primary_assembly.annotation.noheader.genelist.snRNA_snoRNA
```

```python
!python filter_snRNA_mapping_output.py -i FtoKO.dedeup.bam -o FtoKO.csv --db ./Metadata/snRNA_mouse_brain.database.txt --genelist  ./Metadata/gencode.vM35.primary_assembly.annotation.noheader.genelist.snRNA_snoRNA
```
